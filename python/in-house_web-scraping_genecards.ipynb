```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
import pandas as pd 
import re
import numpy as np

GENELIST = '276ddr' #'mutated_gene_21', 'DDRD_assay_42', 'HR_PARP_132'
PATH = '/Users/ywhsiao/Desktop/Research/BRCAness/public_data/00.genesets_from_genecard/'
myFile = pd.read_csv(str(GENELIST)+'.csv', sep=',', header=None)
print(myFile[0].shape[0])

chromedriver = "/Users/ywhsiao/miniconda3/envs/py3.7/bin/chromedriver"
 
chromeOptions = webdriver.ChromeOptions()
prefs = {"download.default_directory": "/Users/ywhsiao/Desktop/Research/BRCAness/genesets_from_genecard"}
chromeOptions.add_experimental_option("prefs", prefs)
chromeOptions.add_argument('--headless')

os.chdir(PATH)

b = []
print(b)
# hg38
for k in range(0,myFile[0].shape[0],1):
    pageurl = "https://www.genecards.org/cgi-bin/carddisp.pl?gene="+str(myFile.iloc[k][0])
    print(pageurl)
    driver = webdriver.Chrome(executable_path=chromedriver, options=chromeOptions)
    driver.get(pageurl)

    pos_hg38 = re.split(':|-',driver.find_elements_by_xpath('//div[@class="col-xs-12 col-sm-4"]/dl[@class="dl-inline"]/dt')[0].text)
    pos_hg38[1] = pos_hg38[1].replace(',', '')
    pos_hg38[2] = pos_hg38[2].replace(',', '')
    size_hg38 = re.split('\s',driver.find_elements_by_xpath('//div[@class="col-xs-12 col-sm-4"]/dl[@class="dl-inline"]/dd')[1].text)[0]
    size_hg38 = size_hg38.replace(',', '')
    pos_hg38.append(size_hg38)
    print(pos_hg38)

    b.append(pos_hg38)
    print(b)
    driver.close()
my_df = pd.DataFrame(b, columns = ["chr", "start", "end", "size"])
my_df.set_index(myFile[0], inplace=True)
my_df.head()
my_df.to_csv(str(GENELIST)+'_loc_with_genename.txt', sep='\t', index=True)
my_df.to_csv(str(GENELIST)+'_loc.txt', sep='\t', index=False)
my_df1 = my_df.drop(['size'], axis=1)
my_df1.to_csv(str(GENELIST)+'.txt', sep='\t', header = False, index=False)
```
